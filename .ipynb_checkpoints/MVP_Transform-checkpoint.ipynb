## Camera direction


```python
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

camera = Camera(position=[0.1, 0, 1])
n = camera.lookAt - camera.position
u = np.cross(camera.up, n)
v = np.cross(n, u)
print('uvn: ', u, v, n)

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.quiver(*camera.position, u, v, n, length=1, color=['r', 'g', 'b'])

ax.set_axis_on()
ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)
ax.set_zlim(-2, 2)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Camera Direction')
```

Camera Matrix

```python
cameraMatrix = camera.viewMatrix()
print(cameraMatrix)
```

```python
triangle = [[0, 0.5, 0], [0.5, 0, 0], [-0.5, 0, 0]]
xs, ys, zs = list(zip(*triangle))

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)
ax.set_zlim(-2, 2)

ax.add_collection3d(Poly3DCollection([triangle]))
ax.quiver(0, 0, 0, u, v, n, length=1, color=['r', 'g', 'b'])
```

### apply camera matrix on triangle
rotate camera position will apply the matrix to vertices

```python
convertedTriangle = []
#print(cameraMatrix)
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)
ax.set_zlim(-2, 2)

for vec3 in triangle:
    vec4 = np.array([*vec3, 1])

    vec4 = (cameraMatrix * vec4.reshape(4, 1)).reshape(4).tolist()[0]

    convertedTriangle.append(vec4[:3])

print(convertedTriangle)
ax.add_collection3d(Poly3DCollection([convertedTriangle]))
ax.quiver(0, 0, 0, u, v, n, length=1, color=['r', 'g', 'b'])
```
